```python
BOOKS_DATA_PATH = "../data/processed_data.csv"
REVIEWS_DATA_PATH = "../data/reviews.csv"
USERS_NUM = 25
```

```python
books_data = pd.read_csv(BOOKS_DATA_PATH, sep=";")
books_data.head()
```

```python
user_data = df = pd.DataFrame(None, index=np.arange(USERS_NUM), columns=books_data.index) 
user_data.head()
```

```python
books_data = books_data.loc[:, ["author", "book_title"]]
most_popular_authors = books_data["author"].value_counts().nlargest(USERS_NUM)
users = []
for key, val, x in zip(most_popular_authors.keys(), most_popular_authors.values, range(USERS_NUM)):
    user_data.iloc[x, books_data.where(books_data["author"] == key).dropna().index.to_list()] = np.random.randint(low = 3,high=6,size=val)
    user_data.iloc[x, 0:9] = 1
user_data = user_data.sub(user_data.mean(axis=1), axis=0).fillna(0)
user_data.head()
```

```python
user_data.to_csv(REVIEWS_DATA_PATH, sep=";")
```
